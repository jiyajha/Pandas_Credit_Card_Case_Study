### 1. In the above dataset, 
#### a. In case age is less than 18, replace it with mean of age values. 

```python
customer_acqusition["Age"] = np.where((customer_acqusition["Age"] < 18), customer_acqusition.Age.mean(), customer_acqusition.Age)
customer_acqusition
```

 ####  b. In case spend amount is more than the limit, replace it with 50% of that customer’s limit. (customer’s limit provided in acquisition table is the per transaction limit on his card)

```python
customer_acqusition_spend = pd.merge(left = customer_acqusition, right = customer_spend, on = "Customer", how = "right", indicator = True)
```

```python
customer_acqusition_spend.head()
```

```python
customer_acqusition_spend["Amount"] = np.where((customer_acqusition_spend.Amount > customer_acqusition_spend.Limit), (customer_acqusition_spend.Limit/2), customer_acqusition_spend.Amount)
```

```python
customer_acqusition_spend.head()
```

 #### c. Incase the repayment amount is more than the limit, replace the repayment with the limit

```python
customer_acqusition_repayment = pd.merge(left = customer_acqusition, right = customer_repayment, on = "Customer", how = "right", indicator = True)
```

```python
customer_acqusition_repayment.head()
```

```python
customer_acqusition_repayment["Amount"] = np.where((customer_acqusition_repayment.Amount > customer_acqusition_repayment.Limit), (customer_acqusition_repayment.Limit), customer_acqusition_repayment.Amount)
```

```python
customer_acqusition_repayment.head()
```

 #### e. If the monthly rate of interest is 2.9%, what is the profit for the bank for each month? (Profit is defined as interest earned on Monthly Profit. Monthly Profit = Monthly repayment – Monthly spend. Interest is earned only on positive profits and not on negative amounts)

```python
bank_monthly_repayment = customer_acqusition_repayment.copy()
bank_monthly_repayment["unique_identify"] = bank_monthly_repayment.Transaction_date.dt.month.astype("str")+"-"+bank_monthly_repayment.Transaction_date.dt.year.astype("str")
bank_monthly_repayment = bank_monthly_repayment.groupby(["unique_identify"])[["Amount"]].sum().add_prefix("Bank_Monthly_Repayment_")
bank_monthly_repayment.index.names = ["Month"]
bank_monthly_repayment.head()
```

```python
bank_monthly_spend = customer_acqusition_spend.copy()
bank_monthly_spend["unique_identify"] = bank_monthly_spend.Transaction_date.dt.month.astype(str)+"-"+bank_monthly_spend.Transaction_date.dt.year.astype(str)
bank_monthly_spend = bank_monthly_spend.groupby(["unique_identify"])[["Amount"]].sum().add_prefix("Bank_Monthly_Spend_")
bank_monthly_spend.index.names = ["Month"]
bank_monthly_spend.head()
```

```python
bank_monthly_profit = pd.merge(left = bank_monthly_repayment, right = bank_monthly_spend, on = "Month", how = "outer", indicator = True)
bank_monthly_profit["Profit"] = bank_monthly_profit.Bank_Monthly_Repayment_Amount - bank_monthly_profit.Bank_Monthly_Spend_Amount
bank_monthly_profit = bank_monthly_profit.loc[bank_monthly_profit["Profit"] > 0]
bank_monthly_profit["Monthly_Profit"] = (bank_monthly_profit["Profit"] * 2.9 )/ 100
```

#### c. Comparison of monthly spend for each product (look for any seasonality     that exists in terms of spend)

```python
from statsmodels.tsa.seasonal import seasonal_decompose

monthly_spend_product = customer_acqusition_spend.copy()
monthly_spend_product["Monthly_Identifier"] = monthly_spend_product.Transaction_date.dt.month.astype(str)+"-"+monthly_spend_product.Transaction_date.dt.year.astype(str)
monthly_spend_product = monthly_spend_product.groupby(["Product", "Monthly_Identifier"])[["Amount"]].sum().add_prefix("Total").unstack()
monthly_spend_product = monthly_spend_product.T.reset_index()
monthly_spend_product = monthly_spend_product.set_index("Monthly_Identifier").iloc[:, range(1,4)]
monthly_spend_product["Date"] = pd.to_datetime(monthly_spend_product.index)
monthly_spend_product = monthly_spend_product.set_index("Date")
monthly_spend_product = monthly_spend_product.fillna(monthly_spend_product.Silver.mean())
```

```python
monthly_spend_product
```

```python
# Let's see if there is any seasonality in Gold Spend
Result = seasonal_decompose(monthly_spend_product.Gold, model = "additive", freq = 1)
Result.seasonal.plot()
plt.show()  
## So, there is no seasonality in Gold Spend
```

```python
# Let's see if there is any seasonality in platinum spend
result = seasonal_decompose(monthly_spend_product.Platimum, model = "additive", freq = 1)
result.seasonal.plot()
plt.show()
## So, there is no seasonality in platinum spend
```

```python
# Let's see if there is any seasonality in silver spend
res = seasonal_decompose(monthly_spend_product.Silver, model = "additive", freq = 1)
res.seasonal.plot()
plt.show()
## So, there is no seasonality in the silver spend
```

### 5. Write user defined PYTHON function to perform the following analysis: 
### You need to find top 10 customers for each city in terms of their repayment amount by different products and by different time periods i.e. year or month. The user should be able to specify the product (Gold/Silver/Platinum) and time period (yearly or monthly) and the function should automatically take these inputs while identifying the top 10 customers.


```python
def top_ten_customer(dataframe):
    dataframe["Yearly_Identifier"] = dataframe.Transaction_date.dt.year.astype(str)
    dataframe["Monthly_Identifier"] = dataframe.Transaction_date.dt.month.astype(str)+"-"+dataframe.Transaction_date.dt.year.astype(str)
    dataframe = dataframe.groupby(["City", "Product", "Customer", "Monthly_Identifier", "Yearly_Identifier"])[["Amount"]].sum().add_prefix("Total_")
    dataframe = dataframe.sort_values("Total_Amount", ascending = False).groupby("City").head(10)
    return dataframe
```
